```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of x and y coordinates
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Create the z values using a mathematical function
# Here we're using a combination of sine waves
Z = np.sin(np.sqrt(X**2 + Y**2)) + np.sin(X*0.5)*np.cos(Y*0.5)

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface with a color map
surface = ax.plot_surface(X, Y, Z, cmap='viridis', 
                         edgecolor='none')

# Add a color bar to show the mapping between colors and z-values
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=5)

# Set labels for each axis
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Add a title
plt.title('3D Surface Plot of sin((x + y)) + sin(0.5x)cos(0.5y)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
fig = plt.figure(figsize=(15, 6))

# 

# 2D
ax1 = fig.add_subplot(121)
t = np.linspace(0, 4*np.pi, 200)
y = np.sin(t)

# 2D
ax1.plot(t, y, 'b-', linewidth=2)
ax1.set_title('2D Sine Wave')
ax1.set_xlabel('t')
ax1.set_ylabel('sin(t)')
ax1.grid(True)

# 2D(a)
fig.text(0.25, 0.02, '(a)', ha='center', va='center', fontsize=10)

# 2D
ax1.text(0.98, 0.95, r'$y = \sin(t)$' + '\n' + r'$t \in [0, 4\pi]$',
         transform=ax1.transAxes,
         bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'),
         verticalalignment='top',
         horizontalalignment='right',
         fontsize=10)

# 3D
ax2 = fig.add_subplot(122, projection='3d')

# 
t = np.linspace(0, 6*np.pi, 200)
x = np.cos(t)
y = np.sin(t)
z = t/3

# 3D
ax2.plot(x, y, z, 'b-', linewidth=2)
ax2.set_title('3D Helix')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

# 3D(b)
fig.text(0.75, 0.02, '(b)', ha='center', va='center', fontsize=10)

# 
equations =r'$x = \cos(t)$' + '\n' + \
           r'$y = \sin(t)$' + '\n' + \
           r'$z = t/3$' + '\n' + \
           r'$t \in [0, 6\pi]$'

ax2.text2D(1.2, 0.5, equations,
           transform=ax2.transAxes,
           bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'),
           verticalalignment='center',
           fontsize=10)

# 3D
ax2.view_init(elev=20, azim=45)

# 
plt.subplots_adjust(right=0.85, bottom=0.15)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
plt.figure(figsize=(10, 8))

# 
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)

#  F(x,y) = x + y - 1
#  F(x,y) = 0 (x,y)
F = X**2 + Y**2 - 1

# 
# 
levels = np.linspace(-1, 1, 15)
contour = plt.contour(X, Y, F, levels=levels, cmap='coolwarm')

#  F(x,y) = 0 
plt.contour(X, Y, F, levels=[0], colors='black', linewidths=2)

# 
plt.text(1, 1, r'$F(x,y) > 0$', fontsize=12)
plt.text(-0.25, 0, r'$F(x,y) < 0$', fontsize=12)
plt.text(0.7, 0, r'$F(x,y) = x^2 + y^2 - 1 = 0$', 
         bbox=dict(facecolor='white', edgecolor='none', alpha=0.7),
         fontsize=12)

# 
plt.clabel(contour, inline=True, fontsize=8, fmt='%1.1f')

# 
# plt.title('Figure 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.3)
plt.axis('equal')  # 

# 
# plt.figtext(0.5, 0.02, 
#            '(a) Implicit representation of a circle:\n' + 
#            r'The curve is defined by $F(x,y) = x^2 + y^2 - 1 = 0$',
#            ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2 - 1

# 
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 
surface = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# z=0
xx, yy = np.meshgrid([-2, 2], [-2, 2])
zz = np.zeros_like(xx)
ax.plot_surface(xx, yy, zz, alpha=0.3, color='gray')

# 
theta = np.linspace(0, 2*np.pi, 100)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
circle_z = np.zeros_like(theta)
ax.plot(circle_x, circle_y, circle_z, 'r-', linewidth=3, label='Intersection')

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 
plt.title('Intersection of Surface F(x,y)=x+y-1 with F(x,y)=0 plane')

# 
ax.text(2, 0, 2, 'F(x,y)=x + y - 1', color='blue', fontsize=12)
ax.text(2,-1, 0, 'F(x,y)=0', color='black', fontsize=12)
ax.text(1, 0, 0, 'F(x,y)=x + y - 1 = 0', color='red', fontsize=12)

# 
fig.colorbar(surface)

# 
ax.legend()

# 
ax.view_init(elev=-10, azim=45)

# 
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skimage.measure import marching_cubes

# 3D
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
z = np.linspace(-3, 3, 100)
X, Y, Z = np.meshgrid(x, y, z)

# 
R = 2  # 
r = 0.5  # 

# 
F = (X**2 + Y**2 + Z**2 + R**2 - r**2)**2 - 4*R**2*(X**2 + Y**2)

# 
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# F=0
verts, faces, _, _ = marching_cubes(F, level=0, step_size=3)

# 
verts = verts * 6 / 100 - 3

# 
ax.plot_trisurf(verts[:, 0], verts[:, 1], verts[:, 2], 
                triangles=faces, cmap=plt.cm.Spectral, alpha=0.8)

# 
equation = r"$F(x,y,z)=(x^2 + y^2 + z^2 + R^2 - r^2)^2 - 4R^2(x^2 + y^2)$"
# 
ax.text2D(0.05, 0.95, equation + "\nR = 2, r = 0.5" + "\nlevel=0", transform=ax.transAxes, linespacing=1.5)

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
# ax.set_title('Implicit Surface: Torus')

# 
ax.view_init(elev=30, azim=45)

# 
ax.set_box_aspect([1,1,1])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
u = np.linspace(0, 2*np.pi, 30)  # 
v = np.linspace(0, 2*np.pi, 30)
U, V = np.meshgrid(u, v)

# 
R = 2  # 
r = 0.5  # 

# 
X = (R + r*np.cos(V)) * np.cos(U)
Y = (R + r*np.cos(V)) * np.sin(U)
Z = r * np.sin(V)

# 
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# 
ax.plot_wireframe(X, Y, Z, color='black', linewidth=0.5)

# 
equation = r"$(x^2 + y^2 + z^2 + R^2 - r^2)^2 = 4R^2(x^2 + y^2)$"
ax.text2D(0.05, 0.95, equation + "\nR = 2, r = 0.5", transform=ax.transAxes)

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Implicit Surface: Torus (Wireframe)')

# 
ax.view_init(elev=30, azim=45)

# 
ax.set_box_aspect([1,1,1])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skimage.measure import marching_cubes

# 3D
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
z = np.linspace(-3, 3, 100)
X, Y, Z = np.meshgrid(x, y, z)

# 
R = 2  # 
r = 0.5  # 

# 
F = (X**2 + Y**2 + Z**2 + R**2 - r**2)**2 - 4*R**2*(X**2 + Y**2)

# 2x2
fig = plt.figure(figsize=(18, 22))

# level
levels = [0, 3, -3, 15]
labels = ['a', 'b', 'c', 'd']

# 
fig.suptitle('Torus Implicit Surface at Different Levels', fontsize=20, y=0.95)

for i, (level, label) in enumerate(zip(levels, labels)):
    # 
    ax = fig.add_subplot(2, 2, i+1, projection='3d')
    
    # 
    verts, faces, _, _ = marching_cubes(F, level=level, step_size=3)
    
    # 
    verts = verts * 6 / 100 - 3
    
    # 
    surf = ax.plot_trisurf(verts[:, 0], verts[:, 1], verts[:, 2],
                          triangles=faces, cmap=plt.cm.Spectral, alpha=0.8)
    
    # 
    equation = r"$F(x,y,z)=(x^2 + y^2 + z^2 + R^2 - r^2)^2 - 4R^2(x^2 + y^2)$"
    ax.text2D(0.05, 0.95, 
              f"{equation}\nR = 2, r = 0.5", 
              transform=ax.transAxes, 
              linespacing=1.5,
              fontsize=16)
    
    # 
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    # 
    ax.view_init(elev=30, azim=45)
    
    # 
    ax.set_box_aspect([1,1,1])
    
    # 
    plt.figtext(0.25 + (i % 2) * 0.5,  # x 
                0.485 if i < 2 else 0.05,  # y 
                f'({label}) level = {level}',
                ha='center',
                va='center',
                fontsize=18)
    #
    ax.tick_params(axis='both', which='major', labelsize=12)

# 
plt.tight_layout(rect=[0, 0.03, 1, 0.95], h_pad=6)  # 

plt.show()
```
